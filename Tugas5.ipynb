```python
import matplotlib.pyplot as plt
```

```python
print("1.pengeluaran gaji + asuransi setiap perusahaan")
print("")

#A-Xis
a = 300000
b = 20
c = 75000

hasil_a = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan A-Xis : " + str(hasil_a))

#B-Log
a = 360000
b = 14
c = 75000

hasil_b = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan B-Log : " + str(hasil_b))

#C-In
a = 475000
b = 28
c = 75000

hasil_c = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan C-In : " + str(hasil_c))

#D-Namic
a = 580000
b = 42
c = 75000

hasil_d = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan D-Namic : " + str(hasil_d))

#E-Psilon
a = 400000
b = 18
c = 75000

hasil_e = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan E-Psilon : " + str(hasil_e))
```

```python
print("	2. Total Pajak (pertahun) dan Utang setiap perusahaan")

pajak = 0.03
utang_a = 0.08
utang_b = 0.04
utang_c = 0.12
utang_d = 0.11
utang_e = 0.03

laba_a = 32500000
laba_b = 31000000
laba_c = 42000000
laba_d = 51500000
laba_e = 27000000


hasil_a = (pajak * laba_a) + (utang_a * 280000000)
print("Maka total pajak dan utang A adalah : " + str(hasil_a))
hasil_b = (pajak * laba_b) + (utang_b * 310000000)
print("Maka total pajak dan utang B adalah : " + str(hasil_b))
hasil_c = (pajak * laba_c) + (utang_c * 420000000)
print("Maka total pajak dan utang C adalah : " + str(hasil_c))
hasil_d = (pajak * laba_d) + (utang_d * 550000000)
print("Maka total pajak dan utang D adalah : " + str(hasil_d))
hasil_e = (pajak * laba_e) + (utang_e * 340000000)
print("Maka total pajak dan utang E adalah : " + str(hasil_e))
print("")

#Grafik bar
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
perusahaan = ['A-Xis', 'B-Log', 'C-In', 'D-Namic', 'E-Psilion']
no = [23375000, 13330000, 51660000, 62045000, 11010000]
c = ['blue', 'green', 'red', 'yellow', 'red']
plt.title("DATA pajak dan utang Perusahaan")
plt.bar(perusahaan, no, color = c)
plt.show()
print("")

#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis ('equal')
plt.title("Data siswa PPD SMK Al-Madani")
perusahaan = ['A-Xis', 'B-Log', 'C-In', 'D-Namic', 'E-Psilion']
no = [23375000, 13330000, 51660000, 62045000, 11010000]
ax.pie(no,labels=perusahaan,autopct='%1.2f%%')
plt.show()
```

```python
print("5. PENDAPATAN BERSIH PERTAHUN TIAP PERUSAHAAN")

tahun = 12
csr_1 = 0.025
csr_2 = 0.035

laba_a = 32500000
laba_b = 31000000
laba_c = 42000000
laba_d = 51500000
laba_e = 27000000

aset_a = 280000000
aset_b = 310000000
aset_c = 420000000
aset_d = 550000000
aset_e = 340000000

gaji_a = 7500000
gaji_b = 6090000
gaji_c = 15400000
gaji_d = 27510000
gaji_e = 8550000


#A
hasil_a = ( laba_a * tahun ) - csr_1 * aset_a - (gaji_a * tahun) - 23375000
print("Pendapatan bersih PerTahun perusahaan A-Xis Rp." + str(hasil_a))

#B
hasil_b = ( laba_b * tahun ) - csr_2 * aset_b - (gaji_b * tahun) - 13330000
print("Pendapatan bersih PerTahun perusahaan B-Log Rp." + str(hasil_b))

#C
hasil_c = ( laba_c * tahun ) - csr_2 * aset_c - (gaji_c * tahun) - 51660000
print("Pendapatan bersih PerTahun perusahaan C-in Rp." + str(hasil_c))

#D
hasil_d = ( laba_d * tahun ) - csr_2 * aset_d - (gaji_d * tahun) - 62045000
print("Pendapatan bersih PerTahun perusahaan D-Namic Rp." + str(hasil_d))

#E
hasil_e = ( laba_e * tahun ) - csr_2 * aset_e - (gaji_e * tahun) - 11010000
print("Pendapatan bersih PerTahun perusahaan E-psilon Rp." + str(hasil_e))
print('')


#Grafik bar
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
perusahaan = ['A-Xis', 'B-Log', 'C-In', 'D-Namic', 'E-Psilion']
no = [269625000, 274740000, 252840000, 206585000, 198490000]
c = ['blue', 'yellow', 'black', 'pink', 'red']
plt.title("DATA pajak dan utang Perusahaan")
plt.bar(perusahaan, no, color = c)
plt.show()
print("")

#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis ('equal')
plt.title("Data siswa PPD SMK Al-Madani")
perusahaan = ['A-Xis', 'B-Log', 'C-In', 'D-Namic', 'E-Psilion']
no = [269625000, 274740000, 252840000, 206585000, 198490000]
ax.pie(no,labels=perusahaan,autopct='%1.2f%%')
plt.show()
```
